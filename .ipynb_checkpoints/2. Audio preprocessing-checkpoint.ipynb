### Raw waveform and log spectrogram
Let's examine the amplitudes and frequencies in the above sample audio segment

```python
# Function to return a log spectrogram of a given audio array
def log_specgram(audio, sample_rate, window_size=20,
                 step_size=10, eps=1e-10):
    nperseg = int(round(window_size * sample_rate / 1e3))
    noverlap = int(round(step_size * sample_rate / 1e3))
    freqs, times, spec = signal.spectrogram(audio,
                                    fs=sample_rate,
                                    window='hann',
                                    nperseg=nperseg,
                                    noverlap=noverlap,
                                    detrend=False)
    return freqs, times, np.log(spec.T.astype(np.float32) + eps)
```

```python
# Generating log spectrogram of the sample audio file
freqs, times, spectrogram = log_specgram(y, sr)

# Plotting the raw waveform
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1.set_title('Raw wave of ' + file_name)
ax1.set_ylabel('Amplitude')
ax1.plot(np.linspace(0, sr/len(y), sr), y)

# Plotting the spectrogram
ax2 = fig.add_subplot(212)
ax2.imshow(spectrogram.T, aspect='auto', origin='lower', 
           extent=[times.min(), times.max(), freqs.min(), freqs.max()])
ax2.set_yticks(freqs[::16])
ax2.set_xticks(times[::16])
ax2.set_title('Spectrogram of ' + file_name)
ax2.set_ylabel('Freqs in Hz')
ax2.set_xlabel('Seconds');
```

### Mel spectrogram and MFCCs
Let's look at the mel power spectrogram and MFCCs of the sample audio segment

```python
# Sample mel spectrogram

S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)

# Convert to log scale (dB). We'll use the peak power (max) as reference.
log_S = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')
plt.title('Mel power spectrogram ')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

```python
# Extract top 24 Mel-frequency cepstral coefficients
mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=24)

# Add on the first and second deltas
delta_mfcc  = librosa.feature.delta(mfcc)
delta2_mfcc = librosa.feature.delta(mfcc, order=2)

# Plot the MFCCs, first and second deltas
plt.figure(figsize=(12, 6))

plt.subplot(3,1,1)
librosa.display.specshow(mfcc)
plt.ylabel('MFCC')
plt.colorbar()

plt.subplot(3,1,2)
librosa.display.specshow(delta_mfcc)
plt.ylabel('MFCC-$\Delta$')
plt.colorbar()

plt.subplot(3,1,3)
librosa.display.specshow(delta2_mfcc, sr=sr, x_axis='time')
plt.ylabel('MFCC-$\Delta^2$')
plt.colorbar()

plt.tight_layout()
```

## Feature extraction
The features of interest in speaker recognition are the Mel-frequency cepstral coefficients and their first and second derivatives.

```python
# Function to extract features of interest
def extract_features(file_name):
    # Load the audio file in librosa
    y, sr = librosa.load(file_name)
    
    # Create mel spectrogram from raw audio segment
    S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)

    # Convert to log scale (dB), using peak power as reference
    log_S = librosa.power_to_db(S, ref=np.max)

    # extract MFCCs
    mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=24)
    
    # extract MFCC deltas
    mfcc_delta = librosa.feature.delta(mfcc)
    
    # extract MFCC delta deltas
    mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
    
    # average the MFCCs
    mfcc = np.mean(mfcc.T, axis = 0)
    
    # average the deltas
    mfcc_delta = np.mean(mfcc_delta.T, axis = 0)
    
    # average the delta deltas
    mfcc_delta2 = np.mean(mfcc_delta2.T, axis = 0)
    
    return mfcc, mfcc_delta, mfcc_delta2

# Function to parse audio files in the correct directory
def parse_audio_files(parent_dir, sub_dirs, file_ext="*.wav"):
    # Create empty arrays for 80 features and corresponding labels
    features, labels = np.empty((0,72)), np.empty(0)
    
    # Loop through each subdirectory
    for label, sub_dir in enumerate(sub_dirs):
        
        # Loop through each file
        for fn in glob.glob(os.path.join(parent_dir, sub_dir, file_ext)):
            # Try to extract features
            try:
                mfcc, deltas, delta2s = extract_features(fn)
            except Exception as e:
                print("Error encountered while parsing file: ", fn)
                continue
            
            # Restructure into a n_observations by n_features matrix
            ext_features = np.hstack([mfcc,deltas, delta2s])
            features = np.vstack([features, ext_features])
            
            # Get label from path name
            labels = np.append(labels, fn.split('/')[2])
            
    return np.array(features), np.array(labels)
```

```python
# Extract features and labels from audio files in the data/train directory
features, labels = parse_audio_files('data/train', data_labels)
```
